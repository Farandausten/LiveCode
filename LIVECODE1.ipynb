```python
#Menghitung Kapasitas Resistansi
def kapasitas1(i,v):
    R = v/i
    print("Kapasitas dalam satuan ohm adalah", R)
    
def daya1(i,v):
    P = i * v
    print("Jumlah Daya dalam Watt adalah", P)
      
#Membuat Perulangan supaya balik lagi ke awal
x = 0
while x == 0:
    kalimat = str(input("Masukkan kalimat: [kapasitas/daya]")).upper()
    v = int(input("Masukkan Tegangan dalam volt: [V]"))
    i = int(input("Masukkan Kuat Arus dalam volt: [I]"))

    if kalimat == "KAPASITAS":
        print(kapasitas1(i, v))
    if kalimat == "DAYA":
        print(daya1(i, v))
    else:
        print("Anda salah input")
```
