```python
#Menghitung Kapasitas Resistansi
def kapasitas1(i,v):
    R = v/i
    print("Kapasitas dalam satuan ohm adalah", R)
    
def daya1(i,v):
    P = i * v
    print("Jumlah Daya dalam Watt adalah", P)
      
#Membuat Perulangan supaya balik lagi ke awal
x = 0
while x == 0:
    kalimat = str(input("Masukkan kalimat: [kapasitas/daya]")).upper()
    v = int(input("Masukkan Tegangan dalam volt: [V]"))
    i = int(input("Masukkan Kuat Arus dalam volt: [I]"))

    if kalimat == "KAPASITAS":
        print(kapasitas1(i, v))
    if kalimat == "DAYA":
        print(daya1(i, v))
    else:
        print("Anda salah input")
```

```python
#Ini salah satu cara
def hambatan(a,b):
  "Rumus menghitung kapasitas baterai dalam satuan Ohm, R=V/I"
  return a/b

def daya(a,b):
  "Rumus menghitung daya, P=V*I"
  return a*b
```

```python
df['Hambatan'] = df.apply(lambda x: hambatan(x['V_m'], x['Current (A)']), axis=1)
df['Daya'] = df.apply(lambda x: daya(x['V_m'], x['Current (A)']), axis=1)
df
```

```python
#Ini cara kedua
def R(V,I):
    return V/I

def P(V,I):
    return V*I

resist=[]
power =[]
for cyc in df.Cycle.unique():
    V = df["V_m"][df['Cycle']== cyc]
    I = df["Current (A)"][df['Cycle']==cyc]
    if cyc%2 != 0:
        resist.append(R(V,I))
    else:
        power.append(R(V,I))
```
